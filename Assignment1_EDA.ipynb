# Exploratory Data Analysis

# Abstract
### Exploratory Data Analysis(EDA) is the process od analyzing data sets and arrving at conclusions based on the data set through various visualization methods like through displays in graphs, charts, histograms, heatmaps.
In this Assignment we are performing the Exploratory Data Analysis(EDA) Technique to find out the various Independent factors that affect the Dependent variable GDP of a country.

We are using many data visualization techniques like bar graphs, pair plots, dist plots, heat maps and box plot to understand and visualize the data to see the patterns.

Along with this,the standard statistical functions like correlation and regression (along with the regression plot) are performed to find out the strength of the relationship of the Dependent variable GDP with respect to the other Independent variables present in the data set.
Analysis has been performed on the main congtributing factors with respect to their Means and Medians with respect to the Dependent Variable GDP.

The analysis of the main three contributing sectors of an economy the Agriculture, industry and the Service Sector relationship with regards to the mean and median have also been performed with respect to the GDP in the assignment.

The various graphs and charts used are:
Barplots-Which is a bar graph which shows how the top and bottom 33 countries vary with the GDP 
Pair plots-Showing how each variable(GDP, Phones(per 1000) and Service vary individually and across each other for all countries presented in the data.Its basicallly like a scatter plot
Box plot-Shows how the data is spread across and deviated across the median and also showing the outliers with respect to the inter-quartile range
Heat map-Its a map kind diagram that shows the strength of the relationship(mainly the correlation) of each variable with itself and the other variables and the values of the correlation are mapped
Distplot-To draw a histogram and fit the kernel dednsity estimate for a single variable(its a univariate distribution for the data)

### Show the regression plot with the Numeric data variables that have strong correlation with the GDP hence Climate has been ommitted

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20,20))
plt.subplots_adjust(hspace=0.4)

corr_to_gdp = pd.Series()#To generate a 1 Dimensional array
for col in countries.columns.values[2:]:
    if ((col!='GDP ($ per capita)')&(col!='Climate')):
        corr_to_gdp[col] = countries['GDP ($ per capita)'].corr(countries[col])
abs_corr_to_gdp = corr_to_gdp.abs().sort_values(ascending=False)
corr_to_gdp = corr_to_gdp.loc[abs_corr_to_gdp.index]

for i in range(3):
    for j in range(3):
        sns.regplot(x=corr_to_gdp.index.values[i*3+j], y='GDP ($ per capita)', data=countries,
                   ax=axes[i,j], fit_reg=False, marker='.')
        title = 'correlation='+str(corr_to_gdp[i*3+j])
        axes[i,j].set_title(title)
axes[1,2].set_xlim(0,102)
plt.show()
```
